```python
import requests

# Define the GraphQL API URL and API key
url = "https://api.yelp.com/v3/graphql"
api_key = "qdZpSbfYUELFmLSoVd5AjiZgP6Z_Ilc09BfrwMQIu7WrOf9NE2a-qlvbWo5DhJutzBDYFC9TarFj6LYNMyUd2tVgPuRHTjc3NrTp7nKFLfCWz0YppDYC9ZxcgRA5Z3Yx"

# GraphQL query to fetch restaurant data grouped by city and calculate average ratings, limited to 30
query = """
{
  search(term: "Restaurant", location: "San Francisco", limit: 30) {
    business {
      name
      rating
      location {
        city
      }
    }
  }
}
"""

# Set the headers for the API request
headers = {
    "Content-Type": "application/json",  # Specify JSON content type
    "Authorization": f"Bearer {api_key}"  # Add the API key for authorization
}

# Send a POST request to the Yelp API
response = requests.post(url, json={'query': query}, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the response JSON
    data = response.json()
    # Extract the business data from the response
    businesses = data.get("data", {}).get("search", {}).get("business", [])

    # Calculate average ratings by city
    city_ratings = {}
    city_counts = {}
    for business in businesses:
        city = business.get("location", {}).get("city", "Unknown")
        rating = business.get("rating")

        # Aggregate ratings and counts for each city
        if city in city_ratings:
            city_ratings[city] += rating
            city_counts[city] += 1
        else:
            city_ratings[city] = rating
            city_counts[city] = 1

    # Calculate average ratings for each city
    avg_ratings = {city: city_ratings[city] / city_counts[city] for city in city_ratings}

    # Sort cities by average rating in descending order and limit to top 30
    sorted_avg_ratings = sorted(avg_ratings.items(), key=lambda x: x[1], reverse=True)[:30]

    # Print the results
    print("Average star ratings for restaurants in specific cities:")
    for city, avg_rating in sorted_avg_ratings:
        print(f"{city}: {avg_rating:.2f}")
else:
    # Print an error message if the request fails
    print(f"Request failed with status code: {response.status_code}")
    print(response.text)
```

```python
import requests
from datetime import datetime

# Define the GraphQL API URL and API key
url = "https://api.yelp.com/v3/graphql"
api_key = "qdZpSbfYUELFmLSoVd5AjiZgP6Z_Ilc09BfrwMQIu7WrOf9NE2a-qlvbWo5DhJutzBDYFC9TarFj6LYNMyUd2tVgPuRHTjc3NrTp7nKFLfCWz0YppDYC9ZxcgRA5Z3Yx"

# GraphQL query to fetch restaurant hours
query = """
{
  search(term: "Restaurant", location: "San Francisco", limit: 30) {
    business {
      name
      hours {
        open {
          day
          start
          end
        }
      }
    }
  }
}
"""

# Set the headers for the API request
headers = {
    "Content-Type": "application/json",  # Specify JSON content type
    "Authorization": f"Bearer {api_key}"  # Add the API key for authorization
}

# Send a POST request to the Yelp API
response = requests.post(url, json={'query': query}, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the response JSON
    data = response.json()
    # Extract the business data from the response
    businesses = data.get("data", {}).get("search", {}).get("business", [])

    if not businesses:
        print("No businesses found in the API response.")
    else:
        # Process hours of operation and calculate total hours open
        extended_hours_restaurants = []
        for business in businesses:
            name = business.get("name", "Unknown")
            hours = business.get("hours", [])
            if not hours:
                print(f"No hours of operation found for business: {name}")
                continue

            for hour in hours[0].get("open", []):  # Only processing the first hours set
                day = hour.get("day")
                start = hour.get("start")  # Start time in HHmm format
                end = hour.get("end")  # End time in HHmm format

                if start and end:
                    # Convert to datetime objects for calculation
                    start_time = datetime.strptime(start, "%H%M")
                    end_time = datetime.strptime(end, "%H%M")

                    # Calculate the total open hours, considering overnight shifts
                    if end_time < start_time:  # Handle overnight closing
                        end_time = end_time.replace(day=end_time.day + 1)
                    hours_open = (end_time - start_time).seconds / 3600  # Convert to hours

                    if hours_open >= 12:  # Filter for extended hours
                        extended_hours_restaurants.append({
                            "name": name,
                            "day": day,
                            "start": start_time.strftime("%H:%M"),
                            "end": end_time.strftime("%H:%M"),
                            "hours_open": hours_open
                        })

        # Print the results
        if extended_hours_restaurants:
            print("Restaurants with extended opening hours:")
            for restaurant in extended_hours_restaurants:
                print(f"{restaurant['name']} - Day: {restaurant['day']}, Hours: {restaurant['start']} - {restaurant['end']} ({restaurant['hours_open']} hours)")
        else:
            print("No restaurants with extended opening hours found.")
else:
    # Print an error message if the request fails
    print(f"Request failed with status code: {response.status_code}")
    print(response.text)
```

```python
import requests
from datetime import datetime

# Define the GraphQL API URL and API key
url = "https://api.yelp.com/v3/graphql"
api_key = "qdZpSbfYUELFmLSoVd5AjiZgP6Z_Ilc09BfrwMQIu7WrOf9NE2a-qlvbWo5DhJutzBDYFC9TarFj6LYNMyUd2tVgPuRHTjc3NrTp7nKFLfCWz0YppDYC9ZxcgRA5Z3Yx"

# GraphQL query to fetch restaurant hours and ratings
query = """
{
  search(term: "Restaurant", location: "San Francisco", limit: 30) {
    business {
      name
      rating
      review_count
      location {
        city
      }
      hours {
        open {
          day
          start
          end
        }
      }
    }
  }
}
"""

# Set the headers for the API request
headers = {
    "Content-Type": "application/json",  # Specify JSON content type
    "Authorization": f"Bearer {api_key}"  # Add the API key for authorization
}

# Send a POST request to the Yelp API
response = requests.post(url, json={'query': query}, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the response JSON
    data = response.json()
    # Extract the business data from the response
    businesses = data.get("data", {}).get("search", {}).get("business", [])

    if not businesses:
        print("No businesses found in the API response.")
    else:
        # Get the current day and time
        now = datetime.now()
        current_day = now.weekday()  # 0=Monday, 6=Sunday
        current_time = now.strftime("%H%M")  # Format time as HHmm

        # Filter for restaurants currently open
        currently_open = []
        for business in businesses:
            name = business.get("name", "Unknown")
            rating = business.get("rating", 0)
            review_count = business.get("review_count", 0)
            city = business.get("location", {}).get("city", "Unknown")
            hours = business.get("hours", [])
            if not hours:
                continue

            for hour in hours[0].get("open", []):  # Only process the first hours set
                day = hour.get("day")
                start = hour.get("start")  # Start time in HHmm format
                end = hour.get("end")  # End time in HHmm format

                if day == current_day and start and end:
                    if start <= current_time <= end:
                        currently_open.append({
                            "name": name,
                            "city": city,
                            "rating": rating,
                            "review_count": review_count
                        })
                        break  # No need to check further hours for the same day

        # Sort the results by rating and review count, then limit to 30
        currently_open = sorted(
            currently_open,
            key=lambda x: (-x["rating"], -x["review_count"])
        )[:30]

        # Print the results
        if currently_open:
            print("Restaurants that are currently open:")
            for restaurant in currently_open:
                print(f"{restaurant['name']} - {restaurant['city']}, Rating: {restaurant['rating']}, Reviews: {restaurant['review_count']}")
        else:
            print("No restaurants are currently open.")
else:
    # Print an error message if the request fails
    print(f"Request failed with status code: {response.status_code}")
    print(response.text)
```

```python
import requests
from collections import Counter

# Define the GraphQL API URL and API key
url = "https://api.yelp.com/v3/graphql"
api_key = "qdZpSbfYUELFmLSoVd5AjiZgP6Z_Ilc09BfrwMQIu7WrOf9NE2a-qlvbWo5DhJutzBDYFC9TarFj6LYNMyUd2tVgPuRHTjc3NrTp7nKFLfCWz0YppDYC9ZxcgRA5Z3Yx"

# GraphQL query to fetch restaurant data, including the state information
query = """
{
  search(term: "Restaurant", location: "USA", limit: 30) {
    business {
      location {
        state
      }
    }
  }
}
"""

# Set the headers for the API request
headers = {
    "Content-Type": "application/json",  # Specify JSON content type
    "Authorization": f"Bearer {api_key}"  # Add the API key for authorization
}

# Send a POST request to the Yelp API
response = requests.post(url, json={'query': query}, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the response JSON
    data = response.json()
    # Extract the business location data
    businesses = data.get("data", {}).get("search", {}).get("business", [])

    if not businesses:
        print("No businesses found in the API response.")
    else:
        # Count the number of restaurants in each state
        state_counts = Counter()
        for business in businesses:
            state = business.get("location", {}).get("state")
            if state:
                state_counts[state] += 1

        # Sort states by the number of restaurants in descending order and limit to 30
        sorted_states = state_counts.most_common(30)

        # Print the results
        print("Number of restaurants in each state:")
        for state, count in sorted_states:
            print(f"{state}: {count}")
else:
    # Print an error message if the request fails
    print(f"Request failed with status code: {response.status_code}")
    print(response.text)
```
